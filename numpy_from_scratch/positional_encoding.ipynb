```python
import numpy as np
```

```python

def positional_encoding(X):
    d_model = X.shape[-1]

    starting_matrix = np.zeros((X.shape[0], X.shape[1]))

    encoded_matrix = []

    position = 0

    for row in starting_matrix:

        final_row_encode = np.copy(row)

        for i in range(0, len(row), 2):

            final_row_encode[i] = np.sin(np.divide(position, np.power(10000, np.divide(np.multiply(2, i), d_model))))
        
            final_row_encode[i + 1] = np.cos(
                np.divide(
                    position, 
                    np.power(
                        10000, 
                        np.divide(
                            np.multiply(2, i), 
                            d_model
                        )
                    )
                )
            )

        position += 1

        encoded_matrix.append(final_row_encode)

    return X + encoded_matrix
```

```python
X = np.array([
    [1, 0, 1, 0],
    [0, 1, 0, 1],
    [1, 1, 0, 0]
])

print(positional_encoding(X))
```
