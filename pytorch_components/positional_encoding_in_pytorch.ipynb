```python
import torch
import torch.nn as nn
```

```python
class PositionalEncoding(nn.Module):
    def __init__(self, d_model):
        super().__init__()
        self.d_model = d_model

    def forward(self, X):
        starting_matrix = torch.zeros((X.shape[0], X.shape[1]))

        encoded_matrix = []

        position = 0

        for row in starting_matrix:

            final_row_encode = torch.clone(row)

            for i in range(0, len(row), 2):

                div_term = 10000 ** ((2 * i) / self.d_model)

                final_row_encode[i] = torch.sin(torch.tensor(position / div_term))
                final_row_encode[i + 1] = torch.cos(torch.tensor(position / div_term))

            position += 1

            encoded_matrix.append(final_row_encode)

        encoded_matrix = torch.stack(encoded_matrix)
        return X + encoded_matrix
```

```python
torch.manual_seed(42)
pe = PositionalEncoding(d_model=4)

X = torch.tensor([[1, 0, 1, 0],
                [0, 1, 0, 1],
                [1, 1, 0, 0]], dtype=torch.float32)

output = pe(X)
print(output)
print(output.shape)  # torch.Size([3, 4])

```
